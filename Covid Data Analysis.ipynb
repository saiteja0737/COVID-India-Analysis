### How did vaccinations impact active cases over time? (Correlation between increasing vaccines and decreasing active cases â€” combine both datasets)

```python
vaccine_df = vaccine_df.rename(columns = {'Total Doses Administered' : 'Total_Doses_Admin'})
```

```python
# Convert date columns to datetime format
vaccine_df['Vaccination_Date'] = pd.to_datetime(vaccine_df['Vaccination_Date'], dayfirst=True)
covid_df['Date'] = pd.to_datetime(covid_df['Date'])


# Group vaccination data by date 
daily_vaccination = vaccine_df.groupby('Vaccination_Date')['Total_Doses_Admin'].sum().reset_index()

# Group COVID data by date (sum across states)
daily_cases = covid_df.groupby('Date')['Active_Cases'].sum().reset_index()

# Rename columns to match
daily_vaccination.rename(columns={'Vaccination_Date': 'Date'}, inplace=True)

print(daily_cases)
print(daily_vaccination)
```

```python
# Merge both datasets on Date
merged_df = pd.merge(daily_vaccination, daily_cases, on='Date', how='inner')

# Sort by date
merged_df = merged_df.sort_values('Date')
merged_df
```

```python
# Compute daily new vaccinations
merged_df['Daily Vaccinations'] = merged_df['Total_Doses_Admin'].diff().fillna(0)

# Compute daily change in active cases
merged_df['Daily Change in Active Cases'] = merged_df['Active_Cases'].diff().fillna(0)

merged_df
```

```python
plt.figure(figsize=(12, 7))

# Plot vaccinations
sns.lineplot(data=merged_df, x='Date', y='Total_Doses_Admin', label='Total Vaccinations', color='green')

# Plot active cases
sns.lineplot(data=merged_df, x='Date', y='Active_Cases', label='Active Cases', color='red')

plt.title('Vaccination vs Active Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 7))

# Plot daily vaccinations
sns.lineplot(data=merged_df, x='Date', y='Daily Vaccinations', label='Daily Vaccinations', color='green')

# Plot active cases
sns.lineplot(data=merged_df, x='Date', y='Daily Change in Active Cases', label='Daily Change in Active Cases', color='red')

plt.axvline(pd.to_datetime('2021-05-01'), color='blue', linestyle='--', label='Peak Vaccination Start')
plt.axvline(pd.to_datetime('2021-06-01'), color='purple', linestyle='--', label='Drop in Cases')

plt.title('Active Cases vs. Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

* The analysis demonstrates a strong inverse correlation between COVID-19 vaccinations and active cases.
* As vaccination rates increased, active cases gradually declined, highlighting the critical role of immunization in controlling the pandemic.
* The turning point, observed around May 2021, shows that mass vaccination efforts effectively reduced the spread of the virus.
* This study reinforces the importance of widespread vaccination campaigns in mitigating public health crises and provides evidence-based support for vaccination as a key tool in pandemic response strategies.

### Is there a delayed effect of vaccinations on active cases and deaths?

```python
# Aggregate daily confirmed cases and deaths
covid_summary = covid_df.groupby('Date')[['Confirmed', 'Cured', 'Deaths']].sum().reset_index()
covid_summary['Active_Cases'] = covid_summary['Confirmed'] - (covid_summary['Cured'] + covid_summary['Deaths'])

# Aggregate daily vaccinations
vaccine_summary = vaccine_df.groupby('Vaccination_Date')[['Total_Doses_Admin']].sum().reset_index()
vaccine_summary.rename(columns={'Vaccination_Date': 'Date'}, inplace=True)
```

```python
# Merge both datasets
merged_df = pd.merge(covid_summary, vaccine_summary, on='Date', how='inner')

# Compute daily changes
merged_df['Daily Vaccinations'] = merged_df['Total_Doses_Admin'].diff().fillna(0)
merged_df['Daily Change in Deaths'] = merged_df['Deaths'].diff().fillna(0)

# Compute correlation matrix
correlation_deaths = merged_df[['Total_Doses_Admin', 'Daily Vaccinations', 'Deaths', 'Daily Change in Deaths']].corr()

# Plot correlation heatmap
plt.figure(figsize=(8,5))
sns.heatmap(correlation_deaths, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlation Matrix: Vaccinations & Deaths")
plt.show()
```

```python
merged_df['Vaccinations_Lagged_14'] = merged_df['Daily Vaccinations'].shift(14)
merged_df['Vaccinations_Lagged_28'] = merged_df['Daily Vaccinations'].shift(28)

correlation_lagged_deaths = merged_df[['Vaccinations_Lagged_14', 'Vaccinations_Lagged_28', 'Daily Change in Deaths']].corr()
print(correlation_lagged_deaths)
```

* Stronger correlation at a 28-day lag:
A 14-day lag showed a moderate correlation (0.54).
A 28-day lag showed a stronger correlation (0.74).
This suggests that vaccines needed time to take full effect in reducing mortality.
* Gradual decrease in deaths with increasing vaccinations:
While deaths initially increased due to the pandemicâ€™s peak, the correlation matrix suggests that higher vaccinations aligned with a reduction in daily death rates over time.
* Overall trend aligns with expected vaccine impact:
Vaccination rollouts scaled up significantly in response to rising cases and deaths.
As vaccine coverage increased, a decline in daily deaths followed.
The data suggests a delayed but noticeable impact of vaccinations on reducing mortality.
* Our analysis demonstrates a strong association between increasing vaccinations and a subsequent decline in deaths. The correlation is more pronounced with a 28-day lag, indicating that vaccines take time to provide protective immunity at a population level. This trend aligns with the expected impact of vaccinations in mitigating severe outcomes and reducing mortality rates over time

### Visulaize the statewise proportion of active cases and cured cases.
